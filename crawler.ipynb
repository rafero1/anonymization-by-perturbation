```python
cities = [city.replace(' ', '-') for city in cities]
```

```python
class LatLonSpider(scrapy.Spider):
    name = "latlons"
    download_delay = 1.0
    autothrottle_enabled = True

    def start_requests(self):
        urls = [f'https://pt.db-city.com/Brasil--Ceará--{city}' for city in cities]
        # urls = [f'https://pt.db-city.com/Brasil--Ceará--Fortaleza']

        for url in urls:
            yield scrapy.Request(url=url, callback=self.parse)

    def parse(self, response):
        yield {
            'municipio': "".join(response.css('#headmap h1 a::text').extract()).strip(),
            'lat': "".join(response.css('b.latitude::text').extract()).strip(),
            'lon': "".join(response.css('b.longitude::text').extract()).strip(),
        }

process = CrawlerProcess(settings={
    "FEEDS": {
        "data-spider.csv": {"format": "csv"},
    },
})

process.crawl(LatLonSpider)
process.start() # the script will block here until the crawling is finished
```
